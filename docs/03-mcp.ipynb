# MCP(model context prorocol 模型上下文协议 )

MCP是Claude公司推出并开源的，帮助LLM主动获取第三方的数据的新标准。

## MCP 架构和组件

- MCP主机：AI应用或者是AI赋能的IDE，这通常是用户的交互点，MCP主机在使用LLM处理任务时通常需要外部数据或者工具来完成任务。

- MCP客户端：帮助LLM和MCP服务器进行通信。将LLM请求转为MCP服务器可以理解的格式。然后将MCP服务器返回的数据转化为LLM可以理解的格式。

- MCP服务器：为 LLM 提供上下文、数据和功能的外部服务。

- 传输层：对于本地资源，可以适应stdio格式传输。对于远程资源，建议使用SSE来传输。

## MCP和RAG的区别

RAG 侧重于“补充外部知识”以提升回答事实性，通过搜索非结构化数据来增强内容生成；而 MCP 侧重于“标准化的工具与动作连接”，擅长与外部系统进行实时结构化数据交互和自动化执行
